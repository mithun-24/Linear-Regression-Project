# IQR
* Find Q1 and Q3
* Find IQR=Q3-Q1
* Find lower limit and upper limit
* lower_limit=Q1-1.5*IQR
* upper_limit=Q3+1.5*IQR
* Find records / values which are greater than upper limit and lower than lower limit

```python
# find Q1 and Q3
Q1=data['Newspaper'].quantile(0.25)
print("Lower Quartile", Q1)
Q3=data['Newspaper'].quantile(0.75)
print('Upper Quartile', Q3)
```

```python
# find IQR
IQR =Q3-Q1
IQR
```

```python
# find lower and upper limit
lower_limit=Q1-1.5*IQR
print('lower limit is ',lower_limit)
upper_limit=Q3+1.5*IQR
print('upper limit is ',upper_limit)
```

```python
# find values above upper limit
data.loc[data['Newspaper']>upper_limit]
```

```python
len(data.loc[data['Newspaper']>upper_limit])/len(data)
```

```python
# find values below lower limit
data.loc[data['Newspaper']<lower_limit]
```

```python
#How to handle outliers
# Since data is not normal will replace outliers with median
data.loc[data['Newspaper']>upper_limit,"Newspaper"]=data['Newspaper'].median()
data.loc[data['Newspaper']>upper_limit]
```

## Split data into x and y

```python
x=data.drop('Sales',axis=1)
y=data['Sales']
```

# split data for training and testing

# sklearn
It is one of the scientific library which includes all the packeges related to ml ,data cleaning and feature engineering

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=4)
```

```python
x_train.shape
```

```python
x_test.shape
```

# model building

```python
#import linear regression model
from sklearn.linear_model import LinearRegression
#initialize model
model=LinearRegression()
# Train model with x_train and y_train
model.fit(x_train,y_train)
```

```python
# make predictions / test model
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
#slopes / coefficients of x
model.coef_
```

```python
# intercept
model.intercept_
```

# Evaluate the model

```python
### Loss functions to know the difference between actual and predicted values.
### different Error/Loss functions i regression algm are
 # MSE(Mean Squared Error)
 # MAE(Mean Absolute Error)
# RMSE(Root Mean Squared Error)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
# MSE
MSE=mean_squared_error(y_test,y_pred)
MSE
```

```python
# MAE
MAE=mean_absolute_error(y_test,y_pred)
MAE
```

```python
# RMSE
RMSE=np.sqrt(MSE)
RMSE
```

```python
# r2_score
r2_score(y_test,y_pred)
```
